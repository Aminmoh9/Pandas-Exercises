```python
import pandas as pd
```

```python
#Load the data files file3.xlsx and file4.xlsx using pandas.
file3 = pd.read_excel('file3.xlsx')
file4 = pd.read_excel('file4.xlsx')
file3
```

```python
#Print the column names of both files.
print("Columns in file3.xlsx:")
print(file3.columns)
print("\nColumns in file4.xlsx:")
print(file4.columns)
```

```python
#Compare the names and order of columns in these files to ensure consistency with the data dataframe.
file3_columns = file3.columns
file4_columns = file4.columns
print("\nAre the columns in the same order?")
print(file3_columns.equals(file4_columns))
```

```python
#Rename all column names to lowercase to standardize them.
file3.columns = file3.columns.str.lower()
file4.columns = file4.columns.str.lower()
```

```python
print("\nColumns in file3.xlsx after renaming to lowercase:")
print(file3.columns)
print("\nColumns in file4.xlsx after renaming to lowercase:")
print(file4.columns)
```

```python
#Merge the two dataframes (data_file3 and data_file4) into a single data dataframe.
data=pd.concat([file3,file4], axis=0, keys=["fil3","file4"])
data
```

```python
#Merge the two dataframes (data_file3 and data_file4) into a single data dataframe. For expriment
data1=pd.merge(file3,file4,  on='controln',how='outer', suffixes=("_file3","_file4"))
data1
```

```python
#Drop the columns domain and dob.
data.drop(columns=['domain', 'dob'], inplace=True)
data
```

```python
#For expriment
data1.drop(columns=['domain_file3','domain_file4','dob_file3','dob_file4'], inplace=True)
data1
```

```python
#Rearrange the columns so that ic2, ic3, ic4, and ic5 appear before AVGGIFT.
rearranged_columns=['controln','state','gender','hv1','ic1','hvp1','pobc1','pobc2','ic2','ic3','ic4','ic5','avggift','tcode','target_d']
data=data[rearranged_columns]
data.head()
```

```python
#Standardize gender Values 
data['gender']= data['gender'].str.lower().replace({'male':'M', 'female':'F','m': 'M', 'f': 'F'})
print(data['gender'].unique())
```

```python
data['gender'] = data['gender'].replace({'u': 'Unknown', 'j': 'Unknown', 'a': 'Unknown', None: 'Unknown'})
print(data['gender'].unique())
```
