```python
#Load the data files file3.xlsx and file4.xlsx using pandas.
file3 = pd.read_excel('file3.xlsx')
file4 = pd.read_excel('file4.xlsx')

display(file3.head())
display(file4.head())
```

```python
#Rename all column names to lowercase to standardize them.
file3.columns = file3.columns.str.lower().str.strip().str.replace('_', "")
file4.columns = file4.columns.str.lower().str.strip().str.replace('_', "")
```

```python
#Merge the two dataframes (data_file3 and data_file4) into a single data dataframe.
data=pd.concat([file3,file4], axis=0, keys=["file3","file4"])
data
```

```python
print("Lenght file3:",len(file3))
print("Lenght file4:", len(file4))
print("Lenght data:", len(data))
```

```python
#Drop the columns domain and dob.
data.drop(columns=['domain', 'dob'], inplace=True)
data
```

```python
#Rearrange the columns so that ic2, ic3, ic4, and ic5 appear before AVGGIFT.
rearranged_columns=['controln','state','gender','hv1','hvp1','pobc1','pobc2','ic1','ic2','ic3','ic4','ic5','avggift','tcode','targetd']
data=data[rearranged_columns]
data.head()
```

```python
#Standardize gender Values 
data['gender']= data['gender'].str.lower().replace({'male':'M', 'female':'F','m': 'M', 'f': 'F'})
print(data['gender'].unique())
```

```python
data['gender'] = data['gender'].replace({'u': 'Unknown', 'j': 'Unknown', 'a': 'Unknown', None: 'Unknown'})
print(data['gender'].unique())
```

```python
data['state'] = data['state'].str.upper().replace({'TENNESSEE': 'TN'})
print(data['state'].unique())
```
